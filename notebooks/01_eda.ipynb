# AI Personal Finance Advisor - EDA Notebook

This notebook explores the synthetic dataset and visualizes key patterns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../dataset/sample_transactions.csv')
df.head()
```

## Category Distribution

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='category', order=df['category'].value_counts().index)
plt.title('Transaction Count by Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Amount Distribution by Category

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='category', y='amount')
plt.title('Amount Distribution by Category')
plt.xticks(rotation=45)
plt.yscale('log')
plt.tight_layout()
plt.show()
```
